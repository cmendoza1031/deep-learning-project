```python
# Cell 2: Download FedISIC data (this will take ~10-20 minutes first time)
# Flamby will prompt you to accept data use agreement
from flamby.datasets.fed_isic2019 import FedIsic2019

# Download all 4 centers
for center_id in range(4):
    print(f"Downloading center {center_id}...")
    train_ds = FedIsic2019(center=center_id, train=True)
    test_ds = FedIsic2019(center=center_id, train=False)
    print(f"  Train: {len(train_ds)} samples, Test: {len(test_ds)} samples")
```

```python
# Cell 3: Verify data integrity
print("FedISIC data loaded successfully!")
print("Class distribution per client:")

for center_id in range(4):
    ds = FedIsic2019(center=center_id, train=True)
    print(f"Client {center_id}: {len(ds)} training samples")
```
